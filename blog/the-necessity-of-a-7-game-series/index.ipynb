```python
import numpy as np

outcomes = ['W', 'L']


won_series = 0
lost_series = 0


for _ in range(1000):
    
    
    wins = 0
    losses = 0
    
    for game in range(3):
        if np.random.choice(outcomes, p=[0.6, 0.4]) == 'W':
            wins = wins + 1
        else:
            losses = losses + 1
        if (wins == 2) or (losses == 2):         
            break
    if wins == 2:
        won_series = won_series + 1
    else:
        lost_series = lost_series + 1
        
```
